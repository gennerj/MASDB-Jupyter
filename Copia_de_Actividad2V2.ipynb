## FICHERO JSON

En **FORMATO JSON (Javascript Object Notation)** cada estructura entre llaves es una instancia y dentro de la estructura los atributos con su valor están conformados por parejas clave : valor. Cada pareja de clave:valor se separa por comas. Cada instancia/fila se encierra entre llaves. Las claves siempre son cadenas de texto encerradas entre comillas. Si tuvieramos los atributos (nombre, apellidos, edad) el fichero JSON tendría la forma {"nombre": "Juan", "apellido": "Pérez", "edad": 10}. <br />
